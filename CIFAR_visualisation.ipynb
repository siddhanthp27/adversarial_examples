```python
import os
import cPickle
import matplotlib.pyplot as plt
import numpy as np
import tensorflow

%matplotlib inline
```

```python
def unpickle(file):
    with open(file, 'rb') as f:
        dict = cPickle.load(f)
        
    return dict
```

```python
dirname = './cifar/cifar-10-batches-py'
filename = 'test_batch'
```

```python
file = os.path.join(dirname, filename)
```

```python
data_dict = unpickle(file)
```
